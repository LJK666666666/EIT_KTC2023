# EIT数据格式探索

这个笔记本用于探索EIT（电学阻抗断层成像）数据集的格式，包括参考数据（ref.mat）和测量数据（data1.mat）。

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from pathlib import Path

# 设置路径
data_dir = Path('../..') / 'EvaluationData' / 'evaluation_datasets' / 'level1'
ref_path = data_dir / 'ref.mat'
data1_path = data_dir / 'data1.mat'
```

```python
# 加载参考数据
print('=== 参考数据 (ref.mat) ===')
ref_data = sio.loadmat(str(ref_path))
print(f'文件中的变量: {list(ref_data.keys())}')
print()

# 分析参考数据中的变量
if 'Injref' in ref_data:
    Injref = ref_data['Injref']
    print(f'Injref 形状: {Injref.shape}')
    print(f'Injref 数据类型: {Injref.dtype}')
    print(f'Injref 示例数据 (前5列):')
    print(Injref[:, :5])
    print()

if 'Uelref' in ref_data:
    Uelref = ref_data['Uelref']
    print(f'Uelref 形状: {Uelref.shape}')
    print(f'Uelref 数据类型: {Uelref.dtype}')
    print(f'Uelref 数值范围: [{Uelref.min():.6f}, {Uelref.max():.6f}]')
    print(f'Uelref 示例数据 (前5行):')
    print(Uelref[:5, :])
    print()

if 'Mpat' in ref_data:
    Mpat = ref_data['Mpat']
    print(f'Mpat 形状: {Mpat.shape}')
    print(f'Mpat 数据类型: {Mpat.dtype}')
    print(f'Mpat 示例数据 (前5行):')
    print(Mpat[:5, :])
    print()
```

```python
# 加载测量数据
print('=== 测量数据 (data1.mat) ===')
data1 = sio.loadmat(str(data1_path))
print(f'文件中的变量: {list(data1.keys())}')
print()

# 分析测量数据中的变量
if 'Inj' in data1:
    Inj = data1['Inj']
    print(f'Inj 形状: {Inj.shape}')
    print(f'Inj 数据类型: {Inj.dtype}')
    print(f'Inj 示例数据 (前5列):')
    print(Inj[:, :5])
    print()

if 'Uel' in data1:
    Uel = data1['Uel']
    print(f'Uel 形状: {Uel.shape}')
    print(f'Uel 数据类型: {Uel.dtype}')
    print(f'Uel 数值范围: [{Uel.min():.6f}, {Uel.max():.6f}]')
    print(f'Uel 示例数据 (前5行):')
    print(Uel[:5, :])
    print()

if 'Mpat' in data1:
    Mpat_data = data1['Mpat']
    print(f'Mpat (测量数据) 形状: {Mpat_data.shape}')
    print(f'Mpat (测量数据) 数据类型: {Mpat_data.dtype}')
    print(f'Mpat (测量数据) 示例数据 (前5行):')
    print(Mpat_data[:5, :])
    print()
```

```python
# 数据对比分析
print('=== 数据对比分析 ===')
print(f'参考数据与测量数据的Mpat是否相同: {np.array_equal(Mpat, Mpat_data)}')
print(f'参考数据与测量数据的Inj是否相同: {np.array_equal(Injref, Inj)}')
print()

# 计算电压差
if 'Uel' in data1 and 'Uelref' in ref_data:
    deltaU = Uel - Uelref
    print(f'电压差 deltaU 形状: {deltaU.shape}')
    print(f'电压差 deltaU 数值范围: [{deltaU.min():.6f}, {deltaU.max():.6f}]')
    print(f'电压差 deltaU 示例数据 (前5行):')
    print(deltaU[:5, :])
    print()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 数据可视化
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# 参考电压数据热图
if 'Uelref' in ref_data:
    im1 = axes[0, 0].imshow(Uelref, aspect='auto', cmap='viridis')
    axes[0, 0].set_title('参考电压 Uelref')
    axes[0, 0].set_xlabel('测量模式')
    axes[0, 0].set_ylabel('电极对')
    plt.colorbar(im1, ax=axes[0, 0])

# 测量电压数据热图
if 'Uel' in data1:
    im2 = axes[0, 1].imshow(Uel, aspect='auto', cmap='viridis')
    axes[0, 1].set_title('测量电压 Uel')
    axes[0, 1].set_xlabel('测量模式')
    axes[0, 1].set_ylabel('电极对')
    plt.colorbar(im2, ax=axes[0, 1])

# 电压差热图
if 'Uel' in data1 and 'Uelref' in ref_data:
    im3 = axes[0, 2].imshow(deltaU, aspect='auto', cmap='RdBu_r')
    axes[0, 2].set_title('电压差 deltaU = Uel - Uelref')
    axes[0, 2].set_xlabel('测量模式')
    axes[0, 2].set_ylabel('电极对')
    plt.colorbar(im3, ax=axes[0, 2])

# 电流注入模式
if 'Injref' in ref_data:
    im4 = axes[1, 0].imshow(Injref, aspect='auto', cmap='coolwarm')
    axes[1, 0].set_title('电流注入模式 Injref')
    axes[1, 0].set_xlabel('注入模式')
    axes[1, 0].set_ylabel('电极')
    plt.colorbar(im4, ax=axes[1, 0])

# 测量模式
if 'Mpat' in ref_data:
    im5 = axes[1, 1].imshow(Mpat, aspect='auto', cmap='Set3')
    axes[1, 1].set_title('测量模式 Mpat')
    axes[1, 1].set_xlabel('测量模式')
    axes[1, 1].set_ylabel('电极对')
    plt.colorbar(im5, ax=axes[1, 1])

# 电压差统计分布
if 'Uel' in data1 and 'Uelref' in ref_data:
    axes[1, 2].hist(deltaU.flatten(), bins=50, alpha=0.7, color='skyblue', edgecolor='black')
    axes[1, 2].set_title('电压差分布')
    axes[1, 2].set_xlabel('电压差值')
    axes[1, 2].set_ylabel('频数')
    axes[1, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 详细数据分析
print('=== 详细数据分析 ===')

# 分析电极数量和测量模式
if 'Injref' in ref_data:
    Nel = Injref.shape[0]  # 电极数量
    N_inj = Injref.shape[1]  # 注入模式数量
    print(f'电极数量: {Nel}')
    print(f'注入模式数量: {N_inj}')
    
    # 统计每个电极的注入次数
    injection_counts = np.sum(Injref != 0, axis=1)
    print(f'每个电极的注入次数: {injection_counts}')

if 'Mpat' in ref_data:
    N_pat = Mpat.shape[1]  # 测量模式数量
    print(f'测量模式数量: {N_pat}')

if 'Uelref' in ref_data:
    N_meas = Uelref.shape[0]  # 测量数量
    print(f'每个注入模式的测量数量: {N_meas}')
    print(f'总测量点数: {N_meas * N_inj}')

# 电压数据统计
if 'Uelref' in ref_data and 'Uel' in data1:
    print(f'参考电压均值: {Uelref.mean():.6f}')
    print(f'参考电压标准差: {Uelref.std():.6f}')
    print(f'测量电压均值: {Uel.mean():.6f}')
    print(f'测量电压标准差: {Uel.std():.6f}')
    print(f'电压差均值: {deltaU.mean():.6f}')
    print(f'电压差标准差: {deltaU.std():.6f}')

print('数据格式总结：')
print('- Injref: 电流注入模式矩阵，形状为 (电极数, 注入模式数)')
print('- Uelref: 参考电压测量矩阵，形状为 (测量数, 注入模式数)')
print('- Mpat: 测量模式矩阵，定义了电压测量方式')
print('- Inj/Uel: 实际测量时的电流注入和电压测量数据，格式与参考数据一致')
```
